```python
import sys
from pathlib import Path
import os

color_dict = {'Thy1-GC6s; Cdh23 (Ahl/ahl)':"#8750A1",'Thy1-GC6s; Cdh23 (ahl/ahl)':'#61C055','(F1) Thy1-GC6s; Cdh23 (Ahl/ahl)':'#6783C2'}

##import 
sys.path.insert(0, os.path.join(os.path.dirname(os.path.abspath("")),"Code","Travis"))
import utils

import utils
from pathlib import Path
import pickle 

home = Path("D:\Travis\Cdh23\Cdh23 Data")
mice = utils.groupData(dir=home, analysis_file_name = "wfstack_3.pkl")
```

```python
###widefield: aggregate dates of experiments and compare to DOB

from datetime import datetime
DOBs = []
wf_dates  = []
wf_dates2 = [] 

for i, mouse in mice.mice.iterrows():
    DOBs.append(datetime.strptime(mouse.DOB, "%m/%d/%Y"))
    wf_dates.append(datetime.strptime(mouse.path.name, "%y%m%d"))
    if not mouse.path_6mo.name == '':
        wf_dates2.append(datetime.strptime(mouse.path_6mo.name, "%y%m%d"))
    else:
        wf_dates2.append(None)
    


mice.mice['DOBdt']=DOBs
mice.mice['Imaging_dt'] = wf_dates
mice.mice['Imaging6mo_dt'] = wf_dates2
mice.mice['Imaging_time'] = mice.mice['Imaging_dt']-mice.mice['DOBdt']
mice.mice['Imaging_time_6mo'] = mice.mice['Imaging6mo_dt']-mice.mice['DOBdt']
```

```python
mice.mice.groupby('Genotype').agg({'Imaging_time':['count','mean','std'],'Imaging_time_6mo':['count','mean','std'] })
```

```python
###two-photon: aggregate dates of experiments and compare to DOB

import responses_analysis
data_path = Path("D:/Travis/Cdh23")
test = utils.findUnanalyzedStacks(data_path, pattern="**/2P/L23/[0-9]*", foldername="all")
twop_mice = responses_analysis.get2Pmice(data_path, test)

twop_mice.head()
```

```python
import pandas as pd
DOBs = []
wf_dates  = []
resp_neurons = []

analysis_path = "response_analysis/isResponsive.csv"
for i, mouse in twop_mice.iterrows():
    DOBs.append(datetime.strptime(mouse.DOB, "%m/%d/%Y"))
    wf_dates.append(datetime.strptime(mouse['2P_path'].name, "%y%m%d"))
    isresponsive = pd.read_csv(mouse['2P_path']/ analysis_path)
    resp_neurons.append(isresponsive.isResponsive.value_counts().loc[True])

twop_mice['responsive_neurons'] = resp_neurons
twop_mice['DOBdt']=DOBs
twop_mice['Imaging_dt'] = wf_dates
twop_mice['Imaging_time'] = twop_mice['Imaging_dt']-twop_mice['DOBdt']
twop_mice.groupby('Genotype').agg({'Imaging_time':['count','mean','std','max'],'responsive_neurons':['count','sum','mean','std']})
```
